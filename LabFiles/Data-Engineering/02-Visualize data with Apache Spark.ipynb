## Lab1__02-Visualize data with Apache Spark

```python
%%sql
SELECT * FROM lakehouse.salesorders
---------------------------------------
--View results - Table, chart
---------------------------------------
```

```python

sqlQuery = "SELECT CAST(MONTH(OrderDate) AS INT) AS OrderMonth, \
                SUM((UnitPrice * Quantity) + Tax) AS GrossRevenue \
            FROM salesorders \
            GROUP BY CAST(MONTH(OrderDate) AS INT) \
            ORDER BY OrderMonth"
df_spark = spark.sql(sqlQuery)
df_spark.show()
```

```python
#######################
#Get started with matplotlib
#######################
from matplotlib import pyplot as plt

# matplotlib requires a Pandas dataframe, not a Spark one
df_sales = df_spark.toPandas()

# Create a bar plot of revenue by year
plt.bar(x=df_sales['OrderMonth'], height=df_sales['GrossRevenue'])

# Display the plot
plt.show()
```

```python
#######################
# matplotlib with Customize
#######################

from matplotlib import pyplot as plt

# Clear the plot area
plt.clf()

# Create a bar plot of revenue by year
plt.bar(x=df_sales['OrderMonth'], height=df_sales['GrossRevenue'], color='orange')

# Customize the chart
plt.title('Revenue by OrderMonth')
plt.xlabel('OrderMonth')
plt.ylabel('Revenue')
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(rotation=45)

# Show the figure
plt.show()
```

```python
#######################
#Get started with seaborn
#######################

import seaborn as sns

# Clear the plot area
plt.clf()

# Create a bar chart
ax = sns.barplot(x="OrderMonth", y="GrossRevenue", data=df_sales)
plt.show()
```

```python
#######################
# seaborn with Customize
#######################

import seaborn as sns

# Clear the plot area
plt.clf()

# Set the visual theme for seaborn
sns.set_theme(style="whitegrid")

# Create a bar chart
ax = sns.barplot(x="OrderMonth", y="GrossRevenue", data=df_sales)
plt.show()
```

```python
import seaborn as sns

# Clear the plot area
plt.clf()

# Create a line chart
ax = sns.lineplot(x="OrderMonth", y="GrossRevenue", data=df_sales)
plt.show()
```
